```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_recall_curve, auc
from sklearn.inspection import permutation_importance
from xgboost import XGBClassifier

RANDOM_STATE = 42

# Загрузка данных
df = pd.read_excel('german_credit_data_clean.xlsx')
print(f"Данные загружены. Размер: {df.shape}")
print("Пример данных:")
display(df.head())

# --- Создание целевой переменной HighRisk ---
# Пороги на основе данных (не медианы!)
credit_threshold = df['Credit amount'].quantile(0.75)  # ~5800
duration_threshold = 36
high_risk_purposes = ['radio/TV', 'furniture/equipment', 'car', 'vacation/others']
low_saving = (df['Saving accounts'] == 'little') & (df['Checking account'] == 'little')
low_job = (df['Job'] == 0)

# Условия риска: если выполняется хоть одно из:
cond1 = df['Credit amount'] > credit_threshold
cond2 = df['Duration'] > duration_threshold
cond3 = df['Purpose'].isin(high_risk_purposes)
cond4 = low_saving
cond5 = low_job

df['HighRisk'] = (cond1 | cond2 | (cond3 & (cond1 | cond2)) | cond4 | cond5).astype(int)

# Проверка распределения
print(f"\nРаспределение HighRisk:")
print(df['HighRisk'].value_counts())
print(f"Доля рискованных клиентов: {df['HighRisk'].mean():.2%}")

# --- Определение признаков ---
numeric_features = ['Age', 'Credit amount', 'Duration']
categorical_features = ['Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']

# Проверка
all_features = numeric_features + categorical_features
assert set(all_features) == set(df.drop(columns=['HighRisk']).columns), "Несоответствие списка признаков!"

# --- Создание препроцессора ---
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
    ],
    remainder='drop'
)

print(f"\nЦелевая переменная 'HighRisk' создана.")
print(f"   Порог Credit amount: {credit_threshold:.0f}")
print(f"   Порог Duration: {duration_threshold} мес.")
print(f"   Рисковые цели: {high_risk_purposes}")
print(f"   Рисковые профили: низкие сбережения или Job=0")
```

```python
# --- Этап 2: Стратифицированное разбиение данных на train / valid / test (60/20/20) ---

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Признаки и целевая переменная (HighRisk уже создана на этапе 1)
X = df.drop(columns=['HighRisk'])
y = df['HighRisk']

# Первое разбиение: выделяем 20% под тест (остаётся 80% для train+valid)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE  # используем константу из Этапа 1
)

# Второе разбиение: из оставшихся 80% выделяем 25% под валидацию (20% от исходных)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.25,  # 0.25 * 80% = 20%
    stratify=y_temp,
    random_state=RANDOM_STATE
)

# Проверка размеров и распределения классов
print("Размеры наборов:")
print(f"Train: {X_train.shape[0]} ({X_train.shape[0] / len(df):.0%})")
print(f"Valid: {X_val.shape[0]} ({X_val.shape[0] / len(df):.0%})")
print(f"Test:  {X_test.shape[0]} ({X_test.shape[0] / len(df):.0%})")

print("\nСтратификация (доля класса 1):")
print(f"Train: {y_train.mean():.3f}")
print(f"Valid: {y_val.mean():.3f}")
print(f"Test:  {y_test.mean():.3f}")

# Визуализация распределения классов
plt.figure(figsize=(10, 5))

datasets = {
    'Train': y_train,
    'Valid': y_val,
    'Test': y_test
}

for i, (name, y_data) in enumerate(datasets.items(), 1):
    plt.subplot(1, 3, i)
    plt.pie(
        y_data.value_counts(),
        labels=['Low Risk', 'High Risk'],
        autopct='%1.1f%%',
        colors=['skyblue', 'salmon']
    )
    plt.title(f'{name}\n(N={len(y_data)})')

plt.suptitle('Распределение классов в train / valid / test')
plt.tight_layout()
plt.show()

print("Разбиение завершено. Теперь можно обучать модели.")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_recall_curve, auc
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt  # <-- добавлен импорт

# --- Создание пайплайнов с использованием preprocessor из Этапа 1 ---
pipe_lr = Pipeline([
    ('preprocessor', preprocessor),  # уже определён в Этапе 1
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])

pipe_dt = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42, max_depth=5))
])

# --- Обучение моделей ---
pipe_lr.fit(X_train, y_train)
pipe_dt.fit(X_train, y_train)

# --- Предсказания на валидации ---
y_val_pred_lr = pipe_lr.predict(X_val)
y_val_prob_lr = pipe_lr.predict_proba(X_val)[:, 1]

y_val_pred_dt = pipe_dt.predict(X_val)
y_val_prob_dt = pipe_dt.predict_proba(X_val)[:, 1]

# --- Функция вычисления метрик ---
def calculate_metrics(y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_prob)
    precision, recall, _ = precision_recall_curve(y_true, y_prob)
    pr_auc = auc(recall, precision)
    return acc, f1, roc_auc, pr_auc

acc_lr, f1_lr, roc_auc_lr, pr_auc_lr = calculate_metrics(y_val, y_val_pred_lr, y_val_prob_lr)
acc_dt, f1_dt, roc_auc_dt, pr_auc_dt = calculate_metrics(y_val, y_val_pred_dt, y_val_prob_dt)

# --- Вывод метрик ---
print("=== Метрики на валидации ===")
print("\nLogisticRegression:")
print(f"  Accuracy: {acc_lr:.4f}, F1: {f1_lr:.4f}, ROC-AUC: {roc_auc_lr:.4f}, PR-AUC: {pr_auc_lr:.4f}")

print("\nDecisionTreeClassifier:")
print(f"  Accuracy: {acc_dt:.4f}, F1: {f1_dt:.4f}, ROC-AUC: {roc_auc_dt:.4f}, PR-AUC: {pr_auc_dt:.4f}")

# --- Проверка дисбаланса ---
pos_rate = y_train.mean()
print(f"\nДоля позитивного класса в train: {pos_rate:.3f} ({pos_rate*100:.1f}%)")

if pos_rate < 0.3:
    print("Дисбаланс: метрика PR-AUC обязательна.")
else:
    print("Дисбаланс отсутствует.")

# --- Сравнение метрик ---
metrics = ['Accuracy', 'F1', 'ROC-AUC', 'PR-AUC']
values_lr = [acc_lr, f1_lr, roc_auc_lr, pr_auc_lr]
values_dt = [acc_dt, f1_dt, roc_auc_dt, pr_auc_dt]

x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x - width/2, values_lr, width, label='LogisticRegression', color='lightgreen')
plt.bar(x + width/2, values_dt, width, label='DecisionTree', color='orange')

plt.xlabel('Метрики')
plt.ylabel('Значение')
plt.title('Сравнение метрик на валидации')
plt.xticks(x, metrics)
plt.ylim(0.8, 1.02)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- Краткий вывод ---
if f1_lr > f1_dt:
    winner = "LogisticRegression"
else:
    winner = "DecisionTreeClassifier"

print(f"\nВывод: {winner} — лучшая baseline-модель по F1.")
```

```python
# --- Этап 4: Продвинутая модель и подбор гиперпараметров ---

from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_recall_curve, auc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # <-- добавлен импорт

# --- Создание пайплайна с XGBoost ---
pipe_xgb = Pipeline([
    ('preprocessor', preprocessor),  # из предыдущих этапов
    ('classifier', XGBClassifier(
        random_state=42,
        eval_metric='logloss'
        # use_label_encoder больше не нужен
    ))
])

# --- Пространство гиперпараметров (5–10 комбинаций) ---
param_distributions = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.05, 0.1, 0.2],
    'classifier__subsample': [0.8, 1.0],
    'classifier__colsample_bytree': [0.8, 1.0]
}

# --- Кросс-валидация ---
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# --- Подбор гиперпараметров ---
random_search = RandomizedSearchCV(
    pipe_xgb,
    param_distributions=param_distributions,
    n_iter=10,  # 10 комбинаций
    scoring='f1',  # можно 'accuracy'
    cv=cv,
    random_state=42,
    n_jobs=-1,
    return_train_score=True
)

random_search.fit(X_train, y_train)

# --- Лучшая модель ---
best_model = random_search.best_estimator_

# --- Предсказания на valid ---
y_val_pred_best = best_model.predict(X_val)
y_val_prob_best = best_model.predict_proba(X_val)[:, 1]

# --- Метрики ---
acc_best = accuracy_score(y_val, y_val_pred_best)
f1_best = f1_score(y_val, y_val_pred_best)
roc_auc_best = roc_auc_score(y_val, y_val_prob_best)
precision, recall, _ = precision_recall_curve(y_val, y_val_prob_best)
pr_auc_best = auc(recall, precision)

# --- Вывод ---
print("\n" + "="*60)
print("РЕЗУЛЬТАТЫ ЛУЧШЕЙ МОДЕЛИ (XGBoost)")
print("="*60)
print(f"Accuracy: {acc_best:.4f}")
print(f"F1-Score: {f1_best:.4f}")
print(f"ROC-AUC:  {roc_auc_best:.4f}")
print(f"PR-AUC:   {pr_auc_best:.4f}")

# --- Сравнение с baseline из Этапа 3 ---
print("\nСравнение с LogisticRegression:")
print(f"  F1: {f1_lr:.4f}")
print(f"  Лучшая модель (XGBoost) F1: {f1_best:.4f}")
if f1_best > f1_lr:
    print("XGBoost превзошла baseline (LogisticRegression).")
else:
    print("XGBoost не превзошла baseline (LogisticRegression).")

# --- Лог результатов ---
results_log = pd.DataFrame(random_search.cv_results_)[[
    'mean_test_score',
    'param_classifier__n_estimators',
    'param_classifier__max_depth',
    'param_classifier__learning_rate'
]].sort_values('mean_test_score', ascending=False)

print("\nТОП-5 конфигураций:")
print(results_log.head())

# --- Визуализация результатов RandomizedSearchCV ---
cv_results = pd.DataFrame(random_search.cv_results_)

plt.figure(figsize=(10, 5))
plt.scatter(cv_results['param_classifier__learning_rate'],
            cv_results['mean_test_score'],
            c=cv_results['param_classifier__max_depth'],
            cmap='viridis',
            s=50)
plt.colorbar(label='max_depth')
plt.xlabel('learning_rate')
plt.ylabel('Средний F1 на CV')
plt.title('Результаты RandomizedSearchCV для XGBoost')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# --- Этап 5: Интерпретация и анализ ошибок ---

from sklearn.inspection import permutation_importance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Permutation Importance ---
perm_importance = permutation_importance(
    best_model, X_val, y_val,
    n_repeats=10,
    random_state=RANDOM_STATE,
    scoring='f1'
)

# Получение имён признаков напрямую из обученного preprocessor
try:
    feature_names = best_model.named_steps['preprocessor'].get_feature_names_out()
except ValueError:
    # Если ошибка — используем имена признаков из X_train (если это DataFrame)
    feature_names = numeric_features + list(
        best_model.named_steps['preprocessor']
        .named_transformers_['cat']
        .get_feature_names_out(categorical_features)
    )

# Убедимся в совпадении длин
min_len = min(len(feature_names), len(perm_importance.importances_mean))
feature_names = feature_names[:min_len]
importance_mean = perm_importance.importances_mean[:min_len]
importance_std = perm_importance.importances_std[:min_len]

# --- Таблица важности ---
importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance_mean': importance_mean,
    'importance_std': importance_std
}).sort_values(by='importance_mean', ascending=False)

print("\n=== ВАЖНОСТЬ ПРИЗНАКОВ (Permutation Importance) ===")
print(importance_df.head(10))

# --- График важности ---
plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df.head(10), x='importance_mean', y='feature', palette='viridis', legend=False)
plt.title('Permutation Importance (Top 10 признаков)')
plt.xlabel('Снижение F1 при перестановке признака')
plt.ylabel('Признак')
plt.tight_layout()
plt.show()

# --- 2. Анализ ошибок ---
print("\nАнализ ошибок на валидационном наборе")

y_val_pred = best_model.predict(X_val)
y_val_prob = best_model.predict_proba(X_val)[:, 1]

errors_df = X_val.copy()
errors_df['y_true'] = y_val.values
errors_df['y_pred'] = y_val_pred
errors_df['y_prob'] = y_val_prob

fp = errors_df[(errors_df['y_true'] == 0) & (errors_df['y_pred'] == 1)]
fn = errors_df[(errors_df['y_true'] == 1) & (errors_df['y_pred'] == 0)]

print(f"\nFP: {len(fp)}, FN: {len(fn)}")

print("\n--- Примеры 3 False Positives (модель ошибочно предсказала HighRisk) ---")
if len(fp) > 0:
    display(fp[['Sex', 'Job', 'Saving accounts', 'Checking account', 'Credit amount', 'Duration', 'Purpose', 'y_prob']].head(3))
else:
    print("Нет False Positives")

print("\n--- Примеры 3 False Negatives (модель пропустила реальный HighRisk) ---")
if len(fn) > 0:
    display(fn[['Sex', 'Job', 'Saving accounts', 'Checking account', 'Credit amount', 'Duration', 'Purpose', 'y_prob']].head(3))
else:
    print("Нет False Negatives")

# --- 3. Группировка ошибок по признаку ---
group_feature = 'Saving accounts'
if group_feature in X_val.columns:
    error_analysis = (
        errors_df
        .assign(error=(errors_df['y_true'] != errors_df['y_pred']).astype(int))
        .groupby(group_feature)['error']
        .agg(['mean', 'count'])
        .rename(columns={'mean': 'error_rate', 'count': 'total'})
        .reset_index()
    )
    print(f"\nЧастота ошибок по признаку '{group_feature}':")
    display(error_analysis)

# --- 4. Выводы ---
print("\nКРАТКИЙ АНАЛИЗ")
if len(fp) + len(fn) == 0:
    print("Модель идеально предсказала на валидации — возможно, переобучение.")
else:
    print("Основные источники ошибок:")
    if len(fn) > 0:
        high_credit_fn = fn['Credit amount'].mean()
        print(f"  - FN: у пропущенных HighRisk средний Credit amount = {high_credit_fn:.0f} — модель недооценивает крупные кредиты при низком Job или хороших сбережениях.")
    if len(fp) > 0:
        low_saving_fp = fp[fp['Saving accounts'] == 'little'].shape[0]
        print(f"  - FP: {low_saving_fp} ошибок у клиентов с 'Saving accounts = little' — модель слишком чувствительна к сбережениям, даже при низкой сумме кредита.")

print("\n  Рекомендации:")
print("  - Добавить interaction: Credit amount × Saving accounts")
print("  - Попробовать class_weight='balanced' в XGBoost")
print("  - Проверить порог классификации (не обязательно 0.5)")
print("  - Для бизнеса: использовать эту модель как скрининг — фильтровать клиентов с 'Saving accounts = little' и 'Duration > 36'.")
```
