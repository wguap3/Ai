```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

# Загружаем CSV с правильным разделителем
df = pd.read_csv("german_credit_data.csv", sep=';')


# Убираем лишний столбец с индексом
df = df.drop(columns=['Unnamed: 0'])

# Первичный просмотр данных
print("Размерность таблицы:", df.shape)
display(df.head())
display(df.tail())

# Информация о типах и пропусках
print(df.info())

# Проверка на дубликаты
print("Количество дубликатов:", df.duplicated().sum())
```

```python
#Числовые признаки vs целевой признак
# Бинарный признак: кредит выше медианы = 1, иначе 0
#Мы берём столбец Credit amount (сумма кредита).
#df['Credit amount'] > df['Credit amount'].median() создаёт логический массив: True, если кредит выше медианы, False — если ниже или равен.
#.astype(int) превращает True → 1, False → 0.
#Итого, мы разделяем выборку на две группы:
#HighCredit = 0 — низкий кредит (≤ медиана)
#HighCredit = 1 — высокий кредит (> медиана)
df['HighCredit'] = (df['Credit amount'] > df['Credit amount'].median()).astype(int)
num_cols = ['Age', 'Duration']

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='HighCredit', y=col, data=df)
    plt.title(f"{col} vs HighCredit")
    plt.show()

    sns.violinplot(x='HighCredit', y=col, data=df)
    plt.title(f"{col} vs HighCredit (violin)")
    plt.show()

    # U-критерий Манна–Уитни
    group0 = df[df['HighCredit']==0][col]
    group1 = df[df['HighCredit']==1][col]
    stat, p = mannwhitneyu(group0, group1)
    print(f"{col}: U={stat}, p-value={p:.4f}")

   #Комментарий:
#Если p < 0.05, значит различие между группами статистически значимо.
#Boxplot показывает медиану, выбросы, а violinplot — плотность распределения.
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Функция для расчета Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Список исходных категориальных признаков (до OHE)
cat_cols = ['Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']

# Расчет Cramér's V для каждого категориального признака относительно HighCredit
for col in cat_cols:
    # Пропуски можно заполнить модой для корректного расчета
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])
    
    v = cramers_v(df[col], df['HighCredit'])
    print(f"Cramér's V ({col} vs HighCredit) = {v:.2f}")

#Значение Cramér’s V близкое к 0 → слабая связь
#Значение ближе к 1 → сильная связь
```

```python
import pandas as pd
df = pd.read_csv("german_credit_data.csv", sep=';')

# Убираем лишний столбец с индексом, если он есть
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Проверяем пропуски
missing_counts = df.isnull().sum()
missing_percent = df.isnull().mean() * 100

missing_df = pd.DataFrame({'missing_count': missing_counts, 'missing_%': missing_percent})
print("Пропуски до обработки:")
display(missing_df)

# Замена пропусков в категориальных признаках модой
for col in ['Saving accounts', 'Checking account']:
    df[col] = df[col].fillna(df[col].mode()[0])

# Проверка пропусков после замены
missing_counts_after = df.isnull().sum()
missing_percent_after = df.isnull().mean() * 100

missing_df_after = pd.DataFrame({'missing_count': missing_counts_after, 'missing_%': missing_percent_after})
print("Пропуски после замены модой:")
display(missing_df_after)
```

```python
# --- Анализ выбросов по IQR ---

num_cols = ['Age', 'Credit amount', 'Duration']

for col in num_cols:
    print(f"\nАнализ выбросов для {col}:")
    
    # Сортируем столбец
    sorted_col = df[col].sort_values()
    
    # Квартели
    Q1 = sorted_col.quantile(0.25)
    Q3 = sorted_col.quantile(0.75)
    IQR = Q3 - Q1
    
    # Границы "усов" boxplot
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Выводим статистику
    print(f"Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
    print(f"Нижняя граница = {lower_bound}, Верхняя граница = {upper_bound}")
    
    # Находим выбросы
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    print(f"Количество выбросов: {outliers.shape[0]}")
    if outliers.shape[0] > 0:
        print(f"Примеры выбросов:\n{outliers.values[:10]}")
    
    # Визуализация
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot для {col} (выбросы по IQR)")
    plt.show()
```
